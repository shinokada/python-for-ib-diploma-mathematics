## Missing Value

<center><img src="image/missing.png"></center>

First, we import sample data which has ?, na, NA, N/A, and no data. We check if the data frame has any null values using `isnull()`. `any()` outputs by columns if there are any null values.

```python
import pandas as pd

df = pd.read_csv('./Data/missing.csv')
display(df)
display(df.isnull())
df.isnull().any()
```

Please not that NaN will replace data item where there is no data.
Comparing the first two tables, ? and na are not recognized as null. So we need to use `na_values` to recognized them as null values. This will change to NAN.

```python
df = pd.read_csv('./Data/missing.csv', na_values=['?','na'])
display(df)
df.isnull().any()
```

## Dealing with NaN values 
### Drop Nan

One way to deal with null values is to drop rows or columns. `dropna()` has the default value of `axis=0`. So `dropna()` and `dropna(axis=0)` outputs the same table. This will drop all rows where there is a null value.

```python
df = pd.read_csv('./Data/missing.csv', na_values=['?','na'])
display(df)
df.dropna()
```

```python
df.dropna(axis=0)
```

`axis=1` will drop columns where there is NaN in the column.

```python
df.dropna(axis=1)
```

### Filling up NaN

We can fill up NaN value with the column mean. The following fill NaN in each column with its column mean.

```python
df = pd.read_csv('./Data/missing.csv', na_values=['?','na'])
display(df)
df=df.fillna(value={'col1':df['col1'].mean(),'col2':df['col2'].mean()})
display(df)
```

You can use median as well.

```python
df = pd.read_csv('./Data/missing.csv', na_values=['?','na'])
display(df)
df=df.fillna(value={'col1':df['col1'].median(),'col2':df['col2'].median()})
display(df)
```

Or any value you think it is appropreate.

```python
df = pd.read_csv('./Data/missing.csv', na_values=['?','na'])
display(df)
df=df.fillna(value={'col1':100,'col2':200})
display(df)
```

# How to deal categorical data

Our data has cat, dog and fox under the col3. We use `get_dummies()` to convert categorical variable to dummy/indicator variables. This will create columns with all categories and add either 0 or 1.

```python
df = pd.read_csv('./Data/missing.csv', na_values=['?','na'])
display(df)
dummies = pd.get_dummies(df['col3'])
display(dummies)
```

We concatenate these dummies to our data.

```python
df = pd.concat([df,dummies],axis=1)
display(df)
```

We don't need col3 any more, so let's drop it.

```python
df.drop('col3', axis=1, inplace=True)
display(df)
```

# Descriptive Statistics

Unless you specify the column, `mean`, `max`, `min`, etc. returns the result of all columns. 

```python
all_mean = df.mean()
all_max = df.max()
all_min = df.min()
all_median = df.median()
all_std = df.std()
all_25_quantile = df.quantile(0.25)
# Print only means
print('The means of all data.')
print(all_mean)
```

As an exercise, print `all_median`, `all_max`.


Let's focus on `radius_mean` in our data. We set a variable called `target`.

```python
df=pd.read_csv('http://bit.ly/2vmLyvU')

target = 'radius_mean'
```

We find the max, min, and range of `radius_mean`.

```python
max = df[target].max()
min = df[target].min()
range =  max - min
print('The max is %.1f, the min is %.1f and the range is %.1f' % (max, min, range))
```

The print command prints out within the bracket. Three `%.1` print with one decimal place and will be replaced with max, min, and range. If you want to know more details, please read this [article](https://www.python-course.eu/python3_formatted_output.php).

Let's find out the mean, median and mode.

```python
mean = df[target].mean()
median = df[target].median()
mode = df[target].mode()
print('The mean is %.1f, the median is %.1f and the mode is %.1f.' % (mean, median, mode))
```

The default value for `quartile` is 0.5 which is the same as median or 50%-tile. You can specify the percentile, `quantile(0.25)`.

```python
# lower quartile, upper quartile, interquartile range
q1 = df[target].quantile(0.25)
q3 = df[target].quantile(0.75)
iqr = q3-q1
print('The lower quartile is %.1f, the upper quartile is %.1f and the interquartile range is %.1f' % (q1, q3, iqr))
```

# Matplotlib

Matplotlib has a lot of flexibility. Here you change the outliers color and symbol and showing three box plots together.

```python
data =[df['radius_mean'],df['texture_mean'],df['texture_worst']]
green_diamond = dict(markerfacecolor='g', marker='D')

fig7, ax7 = plt.subplots()
ax7.set_title('Multiple Samples with Different sizes')
ax7.boxplot(data,flierprops=green_diamond)

plt.show()
```

Matplotlib's [histogram](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html?highlight=hist) has a lot of flexibility. You can add different parameters.

```python
#df[target].hist(bins=100)
plt.hist([df['radius_mean'], 
          df['texture_mean']], 
         histtype='barstacked',
         orientation='horizontal',bins=20)
plt.show()
```
